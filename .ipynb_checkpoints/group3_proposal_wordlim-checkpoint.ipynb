## **<u>Introduction</u>**

### **Background information**

Our knowledge levels increase day to day as we learn about processes and technologies around us. User knowledge allows us to understand a user’s knowledge state and gain from interactions with a particular piece of technology. By understanding user knowledge, analysts can determine how to meet learning needs more effectively in the world of technology.

### **Our question**

By modeling and analyzing factors that impact user knowledge, which factors can be used to most accurately predict knowledge levels of users?

### **Dataset being used**

Our project uses the User Knowledge dataset, which requires classification and clustering strategies to analyze its data. The dataset provides information about students’ knowledge status (UNS) of Electrical DC Machines. The 5 attributes recorded can be divided into two groups based on their relation to goal object materials (STG, study time degree; SCG, repetition number degree; PEG, exam performance) or to objects with goal object (STR, study time degree; LPR, exam performance). Class distribution is also recorded, which may influence the data analysis strategies we use (we will consider using classification or clustering in this case).

## **<u>Preliminary Exploratory Data Analysis</u>**

### **Reading our dataset from the web into R**

```r
# import packages
library(tidyverse)
library(readxl)
library(repr)
library(tidymodels)                                    # Install data.table
library("data.table")  
options(repr.matrix.max.rows = 6)
```

### **Cleaning and wrangling our data into a tidy format**

The Knowledge User dataset is split into training and testing data. Our first step is to extract the training data.

### **Summarizing the data**

**Table 1:** Table of training data from the User Knowledge dataset.

```r
train <- read_excel("data/Training_data.xls") %>% mutate(Class = as_factor(UNS))
```

```r
train
```

**Table 2:** Table of percentage distribution from high to low for each category of UNS.

```r
# explore data
num_obs <- nrow(train)
train %>%
  group_by(Class) %>%
  summarize(
    count = n(),
    percentage = n() / num_obs * 100
  )%>%
arrange(by=desc(percentage))
```

### **Visualizing the data**

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_3 <- train %>%
  ggplot(aes( x= STR, y = LPR, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "STR (the degree of study time of user for related objects with goal object)", 
       y = "LPR (the exam performance of user for related objects with goal object)",
       color = "The knowledge level of user") +
  scale_color_manual(labels = c("very_low","low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_3
```

**Figure 1:** Scatterplot of the relationship between STR and LPR; relationship is unclear.

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_2 <- train %>%
  ggplot(aes(x = STG, y = PEG, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "STG (the degree of study time for goal object materials)", 
       y = "PEG (the exam performance of user for goal objects)",
       color = "The knowledge level of user") +
  scale_color_manual(labels = c("very_low","low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_2
```

**Figure 2:** Scatterplot of the relationship between STG and PEG; relationship is clearer, but there are many outlier data points.

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_4 <- train %>%
  ggplot(aes(x = LPR, y = PEG, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "LPR (the exam performance of user for related objects with goal object)", 
       y = "PEG (the exam performance of user for goal objects)",
       color = "The knowledge level of user") +
  scale_color_manual(labels = c("very_low","low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_4
```

**Figure 3:** Scatterplot of the relationship between LPR and PEG; relationship is clear and we see that the data points are in cluster formation.

## **<u>Methods</u>**

First, we created scatterplots of attribute pairings to visualize their relationships and determine which attributes impact UNS the most. Our next step will be to create a model using training data, and use KNN and k-nearest neighbor techniques to maximize our accuracy. Then, we will use testing data to verify our model and test our chosen attributes.

We have chosen to analyze attributes PEG and LPR, which are both related to user exam performance. Although we initially thought attributes relating to goal object materials would be the most useful attributes for prediction, we realized that PEG and LPR are nicely correlated and will provide a solid foundation for our model.

We will use a histogram to visualize our end results, as we expect our data points will form clusters on a scatterplot; a histogram will display the overlaps in cluster formation clearly.
