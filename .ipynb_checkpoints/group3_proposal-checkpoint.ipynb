### Cleaning and wrangling our data into a tidy format:

The Knowledge User dataset was already split into training and testing sets. Therefore, our first step was to extract the training set for our preliminary exploratory data analysis. From here, we renamed the variable names in the training set to make the data tidy.

### Summarizing the data (exploratory data analysis):

Table 1: insert caption here.

```r
train <- read_excel("data/Training_data.xls") %>% mutate(Class = as_factor(UNS))
```

```r
train
```

Table 2: insert caption here.

```r
# explore data
num_obs <- nrow(train)
train %>%
  group_by(Class) %>%
  summarize(
    count = n(),
    percentage = n() / num_obs * 100
  )%>%
arrange(by=desc(percentage))
```

### Visualizing the data (exploratory data analysis):

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_3 <- train %>%
  ggplot(aes( x= STR, y = LPR, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "STR (The degree of study time of user for related objects with goal object)", 
       y = "LPR (The exam performance of user for related objects with goal object)",
       color = "The knowledge level of user") +
  scale_color_manual(labels = c("very_low","low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_3
```
