## **<u>Introduction</u>**

### **Background information**

Our knowledge levels are increasing day to day as we learn about processes and technologies around us. User knowledge allows us to understand a user’s knowledge state and gain from interactions with a particular piece of technology. By understanding user knowledge, analysts can determine how to meet learning needs more effectively within the world of technology.

### **Our question**

By modeling and analyzing factors that impact user knowledge, which factors can be used to most accurately predict knowledge levels of users?

### **Dataset being used**

Our project uses the User Knowledge dataset, which requires classification and clustering strategies to analyze its data. The dataset provides information about students’ knowledge status (UNS) of Electrical DC Machines.

The 5 attributes recorded can be divided into two groups:

1. Study time degree (STG), repetition number degree (SCG) and exam performance (PEG) as related directly to course materials.
2. Study time degree (STR) and exam performance (LPR) as related to beyond course (supplementary) materials.

Class distribution is also recorded, which may influence the data analysis strategies we use. We will consider using KNN (classification) or k-means clustering in this case.

## **<u>Preliminary Exploratory Data Analysis</u>**

### **Reading our dataset from the web into R**

```r
# import packages
library(tidyverse)
library(readxl)
library(repr)
library(tidymodels)                                    # Install data.table
library("data.table")  
options(repr.matrix.max.rows = 6)
```

### **Cleaning and wrangling our data into a tidy format**

The Knowledge User dataset is split into training and testing data. Our first step is to extract the training data.

### **Summarizing the data**

**Table 1:** Table of training data.

```r
train <- read_excel("data/Training_data.xls") %>% mutate(Class = as_factor(UNS))
```

```r
train
```

**Table 2:** Table of percentage distribution from high to low for each category of UNS.

```r
# explore data
num_obs <- nrow(train)
train %>%
  group_by(Class) %>%
  summarize(
    count = n(),
    percentage = n() / num_obs * 100)%>%
arrange(by=desc(percentage))
```

### **Visualizing the data**

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_3 <- train %>%
  ggplot(aes( x= STR, y = LPR, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "STR (user's study time degree of supplementary course materials)", 
       y = "LPR (user's exam performance on supplementary course materials)",
       color = "Knowledge level of user") +
  scale_color_manual(labels = c("very_low","Low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_3
```

**Figure 1:** Scatterplot of the relationship between STR and LPR; relationship is unclear.

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_2 <- train %>%
  ggplot(aes(x = STG, y = PEG, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "STG (user's study time degree of course materials)", 
       y = "PEG (user's exam performance on course materials)",
       color = "Knowledge level of user") +
  scale_color_manual(labels = c("very_low","Low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_2
```

**Figure 2:** Scatterplot of the relationship between STG and PEG; relationship is clearer, but there are many outlier data points.

```r
options(repr.plot.width=17, repr.plot.height=12)

perim_concav_4 <- train %>%
  ggplot(aes(x = LPR, y = PEG, color = Class)) +
  geom_point(alpha = 0.6, size = 2.5) +
  labs(x = "LPR (user's exam performance on supplementary course materials)", 
       y = "PEG (user's exam performance on course materials)",
       color = "Knowledge level of user") +
  scale_color_manual(labels = c("very_low","Low", "Middle","High"), 
                     values = c("orange2", "steelblue2","green","purple")) +
  theme(text = element_text(size = 15))
perim_concav_4
```

**Figure 3:** Scatterplot of the relationship between LPR and PEG; relationship is clear and we see that the data points are in cluster formation.

## **<u>Methods</u>**

First, we created scatterplots of attribute pairings to visualize their relationships and determine which attributes impact UNS the most. Our next step will be to create a model using training data, using KNN (classification) and possibly k-means clustering if it is more effective. Then, we will use testing data to verify our model and test our chosen attributes.

We have chosen to analyze attributes PEG and LPR, which are both related to user exam performance. Although we initially thought attributes related directly to course materials would be the most useful attributes for prediction, we realized that PEG and LPR are nicely correlated and will provide a solid foundation for our model.

We will use a histogram to visualize our end results, as we expect our data points will form clusters on a scatterplot; a histogram will display the overlaps in cluster formation clearly.

## **<u>Expected Outcomes And Significance</u>**

We predict that PEG and LPR will create an accurate model for predicting knowledge levels of users. Since PEG and LPR are both related to user exam performance, our findings will place importance on maximizing user exam grades. This will involve strengthening user study strategies.
